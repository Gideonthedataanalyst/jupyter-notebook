# LOC AND ILOC FUNCTIONS.

```python
#Loc and Iloc allows us to work with the slicing of rows and columns together.
```

## ILOC

```python
# iloc==>columns can only be considered as indices
# for iloc we cannot refer to columns by their string names
# (start:stop:stepby, start:stop:stepby)
# (Row,Column)
```

```python
#survey_df.iloc[row,column]
survey_df.iloc[32400:32445:5,4:10:1]
```

```python
survey_df.columns
```

```python
survey_df.iloc[32400:32445:5,4::]
```

```python
survey_df.iloc[32400:32445:5,:]
```

## LOC function.

```python
#survey_df.loc[row,column]
survey_df.loc[32400:32445:5, "Year"::]
# loc works with string column names
```

```python
survey_df.loc[32400:32445:5,["Year","Value"]]
```

```python
survey_df.loc[::,"Year":"Variable_name":2]
```

```python
survey_df.loc[::,"Year":"Variable_name":2].sample(12)
```

# Subsetting Data Using some Criteria.

```python
#showing our table only where the year is 2018
survey_df[survey_df.Year==2018]
```

```python
survey_df[survey_df.Year==2018].sample(12)
```

```python
#showing our table only where the value column is C
survey_df[survey_df.Value=='C']
```

```python
#showing our table only where the value column is C
survey_df[survey_df.Value=='C'].describe()
```

```python
#showing our table only where the years fall between 2015 and 2020
survey_df[(survey_df.Year>2015)&(survey_df.Year<2020)]
```

```python
#showing our table only where the years fall between 2015 and 2020
survey_df[(survey_df.Year>2015)&(survey_df.Year<2020)].sample(7)
```
